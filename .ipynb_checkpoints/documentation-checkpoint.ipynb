# Probabilistic Ensembles for Control (ProbECon)

This Python module contains functions and classes to apply probabilistic machine learning techniques to control problems.

It contains standard neural networks and deep ensemble models as well as a constrained trajectory optimization method (constrained iLQR) for solving nonlinear optimal control problems.

In future versions it will also contain model-based reinforcement learning algorithms.

# Creating your own system (based on its state-space model)

Every ProbECon environment (system) inherits from the 'gym' environment class and is therefore compatible with many other available reinforcement learning algorithms.

The base class is 'StateSpaceEnv'. Let us consider the following double integrator:

\begin{align}
\dot x_1 =  x_2 \\
\dot x_2 = u_1
\end{align}

At first let us import the necessary packages:

```python
import numpy as np
import probecon
from probecon.system_models.core import StateSpaceEnv
```

No we have to define the ODE of our system:

```python
def ODE(t, x, u):
    x1, x2 = x
    u1, = u
    return np.array([x2, u1])
```

```python
x0 = np.array([0., 0.]) # initial state
```

```python
double_integrator = StateSpaceEnv(state_dim=2, control_dim=1, ode=ODE, time_step=0.01, init_state=x0)
```

To simulate our system, we have to take 'steps'. This means, we apply a control signal 'u' which is applied to the system for 'time_step' seconds.

```python
double_integrator.reset() # reset to x0
for step in range(1000):
    u = np.array([.5])
    double_integrator.step(u)
```

To show the simulation results simply call the 'plot' method.

```python
fig = double_integrator.plot()
```

The corresponding data is contained in the 'trajectory' dictionairy:

```python
traj = double_integrator.trajectory
```

```python
time = traj['time']
states = traj['states']
controls = traj['controls']
```

# Creating your own system via Lagrangian mechanics

We can also create a system model from the Lagrangian. This automates the process of modelling by hand. The process is really straight forward an there are numerous examples in the package, for example an inverted pendulum system.

```python
from probecon.system_models.pendulum import Pendulum
```

```python
pendulum = Pendulum()
```

The controls of the pendulum are bounded. We can sample uniform randomly from the control space via the 'random_step' method.

```python
pendulum.reset()
for step in range(500):
    pendulum.random_step()
```

```python
fig = pendulum.plot()
```

There is also the option, to animate the system in a seperate window, based on the 'render' method from 'gym'.

```python
pendulum.reset()
for step in range(500):
    pendulum.random_step()
    pendulum.render()
pendulum.close()
```
